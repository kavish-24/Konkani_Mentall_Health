```python
# KONKANI AUDIO TRANSCRIPTION - UPLOAD METHOD
# Record on your phone/computer, then upload here

import subprocess
import sys
print("üîß Setting up...")
subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "--upgrade", "google-generativeai"])

import os
import google.generativeai as genai
from google.colab import userdata, files

print("‚úì Ready!\n")

# Get API key
try:
    api_key = userdata.get('Gemini_Key')
    print("‚úì API key loaded\n")
except:
    api_key = input("Enter your Gemini API key: ")

# Configure Gemini
genai.configure(api_key=api_key)
model = genai.GenerativeModel('models/gemini-2.5-flash')

print("="*60)
print("üéôÔ∏è  KONKANI AUDIO TRANSCRIPTION")
print("="*60)
print("\nüì± INSTRUCTIONS:")
print("1. Record Konkani audio on your phone/computer")
print("   - Android: Voice Recorder app")
print("   - iPhone: Voice Memos app")
print("   - Windows: Voice Recorder")
print("   - Mac: Voice Memos")
print("\n2. Upload the file when prompted below")
print("\n3. Supported formats: .wav, .mp3, .m4a, .webm, .ogg, .aac")
print("="*60)
print()

# Choose script
print("Choose script:")
print("1. Devanagari (‡§ï‡•ã‡§Ç‡§ï‡§£‡•Ä) [DEFAULT]")
print("2. Roman (Latin)")
print("3. Kannada (‡≤ï‡≥ä‡≤Ç‡≤ï‡≤£‡≤ø)")
choice = input("\nEnter 1, 2, or 3 (or press Enter for Devanagari): ").strip()

script_map = {"1": "devanagari", "2": "roman", "3": "kannada", "": "devanagari"}
SCRIPT = script_map.get(choice, "devanagari")

print(f"\n‚úì Selected: {SCRIPT.upper()}\n")

# Prompt templates
prompts = {
    "devanagari": "Transcribe this Konkani audio accurately in Devanagari script (‡§ï‡•ã‡§Ç‡§ï‡§£‡•Ä). Only provide the transcription text, no commentary.",
    "roman": "Transcribe this Konkani audio accurately in Roman/Latin script. Only provide the transcription text, no commentary.",
    "kannada": "Transcribe this Konkani audio accurately in Kannada script (‡≤ï‡≥ä‡≤Ç‡≤ï‡≤£‡≤ø). Only provide the transcription text, no commentary."
}

# Upload files
print("="*60)
print("üìÇ UPLOAD YOUR AUDIO FILE(S)")
print("="*60)
print("Click 'Choose Files' below and select your recordings\n")

uploaded = files.upload()

if not uploaded:
    print("\n‚ö†Ô∏è  No files uploaded. Exiting.")
else:
    print(f"\n‚úì {len(uploaded)} file(s) uploaded!\n")

    transcripts = []

    # Process each file
    for i, filename in enumerate(uploaded.keys(), 1):
        print("="*60)
        print(f"üìç PROCESSING FILE {i}/{len(uploaded)}")
        print("="*60)
        print(f"File: {filename}")

        try:
            # Upload to Gemini
            print("üîÑ Uploading to Gemini API...")
            audio_file = genai.upload_file(filename)
            print("‚úì Upload complete")

            # Transcribe
            print("ü§ñ Transcribing Konkani audio...")
            response = model.generate_content([prompts[SCRIPT], audio_file])
            transcript = response.text

            # Cleanup
            genai.delete_file(audio_file.name)

            # Store result
            transcripts.append({
                'filename': filename,
                'transcript': transcript
            })

            print("‚úì Transcription complete!\n")
            print(f"üìù TRANSCRIPT ({SCRIPT.upper()}):")
            print("-"*60)
            print(transcript)
            print("-"*60)
            print()

        except Exception as e:
            print(f"‚ùå Error: {str(e)}\n")
            transcripts.append({
                'filename': filename,
                'transcript': f"[Error: {str(e)}]"
            })

    # Final summary
    print("\n" + "="*60)
    print("üìã ALL TRANSCRIPTIONS")
    print("="*60)
    for i, result in enumerate(transcripts, 1):
        print(f"\nFile {i}: {result['filename']}")
        print(f"Transcript:\n{result['transcript']}\n")

    # Save to file
    output_file = f'konkani_transcripts_{SCRIPT}.txt'
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(f"Konkani Transcriptions ({SCRIPT.upper()} script)\n")
        f.write("="*60 + "\n\n")
        for i, result in enumerate(transcripts, 1):
            f.write(f"File {i}: {result['filename']}\n")
            f.write(f"Transcript:\n{result['transcript']}\n\n")
            f.write("-"*60 + "\n\n")

    print(f"‚úÖ Saved to: {output_file}")
    print("üì• Downloading...\n")
    files.download(output_file)

    print("‚úì Complete!")
    print("‡§¶‡•á‡§µ ‡§¨‡§∞‡•á‡§Ç ‡§ï‡§∞‡•Ç‡§Ç! (God bless!) üôè")
```
